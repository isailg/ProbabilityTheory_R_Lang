# Conteo de Palabras Reservadas en la Librería Agast.cpp
Cargar el archivo agast.cpp, hacer el conteo de las palabras que coinciden con la lista de palabras reservadas.


```python
Archivo_10kLineas = "/gdrive/MyDrive/NLP/agast.cpp"
```

```python
# Líneas del archivo
def count_lines(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        total_lines = len(lines)
    return total_lines

total_lines = count_lines(Archivo_10kLineas)

print(f"El archivo contiene {total_lines} líneas.")
```

```python
# Contar palabras reservadas
from collections import Counter

def conteo_de_palabras_reservadas(file_path, keywords):
    with open(file_path, 'r') as file:
        content = file.read()

    # Divide el texto en palabras y filtrar solo las palabras reservadas
    words = content.split()
    keyword_counts = Counter(word for word in words if word in keywords)
    return keyword_counts
```

```python
keyword_counts = conteo_de_palabras_reservadas(Archivo_10kLineas, palabrasreservadas_cpp)
```

# Frecuencia de las palabras reservadas encontradas

```python
print("Frecuencia de palabras reservadas en el archivo:")
for keyword, count in keyword_counts.items():
    print(f"{keyword}: {count}")
```

```python
keywords_ordenadas = keyword_counts.most_common() # De mayor a menor
```

# Gráfica de los resultados

## Parece tener una distribución de **Ley de Potencia**


```python
import matplotlib.pyplot as plt

# Separar palabras y frecuencias
words, counts = zip(*keywords_ordenadas)

# Crear la gráfica
plt.figure(figsize=(12, 8))
plt.bar(words, counts, color='skyblue')
plt.xticks(rotation=90)
plt.xlabel("Palabras reservadas de C++", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)
plt.title("Frecuencia de palabras reservadas de C++ en la librería Agast.cpp", fontsize=16)
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
